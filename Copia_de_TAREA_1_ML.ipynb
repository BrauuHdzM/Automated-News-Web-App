**Condicionales:** Las declaraciones condicionales son usadas para ejecutar c√≥digo si una cierta condici√≥n es verdadera.

```python
if a > b:  # Si "a" es mayor que "b"
    print("a es mayor que b")
elif a < b:  # Si "a" es menor que "b"
    print("a es menor que b")
else:  # Si ninguna de las condiciones anteriores es verdadera
    print("a es igual a b")
```

Las estructuras condicionales en Python te permiten realizar diferentes acciones basadas en ciertas condiciones. Esto significa que puedes controlar el flujo de tu programa bas√°ndote en los resultados de las expresiones que eval√∫as.

Hay tres formas b√°sicas de estructuras condicionales:

**if** : Esta es la forma m√°s b√°sica. Eval√∫a una condici√≥n y, si esa condici√≥n es verdadera (True), ejecuta el bloque de c√≥digo que sigue.

```python
x = 10
if x > 0:  # Si x es mayor que 0
    print("x es un n√∫mero positivo")
```

En este caso, como x es de hecho mayor que 0, se imprimir√° "x es un n√∫mero positivo".

Ve que pasa si cambias los valores de x a por ejemplo x=0 o x=-1 , ¬øque corregirias para que x=0 pase como un n√∫mero positivo?


**elif:** Esta es una forma de decir "si la(s) condici√≥n(es) anterior(es) no era(n) verdadera(s), entonces intenta esta condici√≥n". Puedes tener tantos elif en tu c√≥digo como quieras.

```python
x = 10
if x > 100:  # Si x es mayor que 100
    print("x es mayor que 100")
elif x > 50:  # Si x es mayor que 50
    print("x es mayor que 50")
elif x > 0:  # Si x es mayor que 0
    print("x es mayor que 0")
```

Intenta ver que pasa, para otros valores de **x**

**else:** Esta es una forma de decir "si ninguna de las condiciones anteriores era verdadera, entonces haz esto". Solo puedes tener un else en tu bloque de c√≥digo.

```python
x = -10
if x > 0:  # Si x es mayor que 0
    print("x es un n√∫mero positivo")
else:  # Si ninguna de las condiciones anteriores es verdadera
    print("x no es un n√∫mero positivo")
```

En este caso, como $x$ no es mayor que $0$, se imprimir√° "$x$ no es un n√∫mero positivo".

Estas estructuras pueden ser tan simples o tan complejas como necesites. Puedes anidarlas (poner una estructura dentro de otra) para hacer evaluaciones m√°s complejas.

Por ejemplo:

```python
x = 10
y = 20
if x > 0:  # Si x es mayor que 0
    if y > 0:  # Si y es mayor que 0
        print("x y y son n√∫meros positivos")
    else:  # Si y no es mayor que 0
        print("x es un n√∫mero positivo, y no lo es")
else:  # Si x no es mayor que 0
    print("x no es un n√∫mero positivo")
```

En este caso, imprimir√° **"x y y son n√∫meros positivos".**

Problema reto:

Escribe un programa que te regrese los primos del 1 al 100000, recuerda que un n√∫mero es primo si y solo si es divisible entre 1 y el mismo n√∫mero.

Por ejemplos los n√∫meros $2,3,5,7,11,13$ son primos, pero el n√∫mero $15$ no es primo ya que es divisible entre $2$ y $3$


Para ver que un n√∫mero $n$ es primo, solo hay que fijarse que los primos menores que $\sqrt n$, explica en otra celda de texto ¬øpor que?, si puedes has la demostraci√≥n.




```python
import math

def es_primo(n):
    if n <= 1:
        return False  # Los n√∫meros menores o iguales a 1 no son primos
    if n <= 3:
        return True  # 2 y 3 son primos
    if n % 2 == 0 or n % 3 == 0:
        return False  # Si es divisible entre 2 o 3, no es primo

    # Comprobar divisibilidad por n√∫meros primos hasta ‚àön
    limite = int(math.sqrt(n)) + 1
    for i in range(5, limite, 2):  # Iterar por n√∫meros impares, comenzando en 5
        if n % i == 0:
            return False
    print("el limite es " + str(limite))
    return True

for numero in range (1, 51):
  if es_primo(numero):
    print(numero)

```
